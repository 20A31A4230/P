```python
from PIL import Image 
import random
end=100
```

```python

def show_board():
    img = Image.open('F:\snakeandladder.jpg')
    img.show()
```

```python
def check_ladder(points):
    if points ==8:
        print("Ladder")
        return 26
    elif points == 21:
        print("Ladder")
        return 82
    elif points == 43:
        print("Ladder")
        return 77
    elif points == 50:
        print("Ladder")
        return 91
    elif points == 62:
        print("Ladder")
        return 96
    elif points == 66:
        print("Ladder")
        return 87
    elif points == 80:
        print("Ladder")
        return 100
    else:
        return points
    
```

```python
def reached_end(points):
    if points == end:
        return True
    else:
        return False
    
```

```python
def play():
    #players Name
    p1_name = input("Enter player1 name")
    p2_name = input("Enter player2 name")
    
    # initial points of Players
    pp1=0
    pp2=0
    turn=0
    
    while(1):
        if turn%2 == 0:  #players1 turn
            
            print(p1_name,' your turn')
            c = input("press 1 to continue and 0 to exit") #for exit or continue
            if c==0:
                input(p1_name, ' scored', pp1)
                print()
                input(p2_name, ' scored', pp2)
                print("Quitting the game...Thanks for playing!!!!")
                break
            
            dice = random.randint(1,6)  #generate random number 
            print("Dice showed ",dice)
            print()

            pp1 = pp1 + dice
            
            pp1 = check_ladder(pp1)
            
            pp1 = check_snake(pp1)
            
            #check if player1 goes beyond the end
            
            if pp1>end:
                pp1=end
            print(p1_name, ' Your Score ',pp1)
            print()

            
            if reached_end(pp1):
                print(p1_name, " Won")
                break
                
        else:
            print(p2_name,' your turn')
            c = input("press 1 to continue and 0 to exit") #for exit or continue
            if c==0:
                input(p1_name, ' scored', pp1)
                print()
                input(p2_name, ' scored', pp2)
                print("Quitting the game...Thanks for playing!!!!")
                break
            
            dice = random.randint(1,6)  #generate random number 
            print("Dice showed ",dice)
            print()
            pp2 = pp2 + dice
            
            pp2 = check_ladder(pp2)
            
            pp2 = check_snake(pp2)
            
            #check if player2 goes beyond the end
            
            if pp2>end:
                pp2=end
            print(p2_name, ' Your Score ',pp2)
            print()
            if reached_end(pp2):
                print(p2_name, " Won")    
                break
                
        turn = turn + 1
        
```

```python
show_board()
play()
```
